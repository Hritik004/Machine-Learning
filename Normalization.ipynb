```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Social_Network_Ads.csv")
```

```python
df
```

```python
sns.kdeplot(df['Age'])
```

```python
color_dict={1:'red',2:'blue'}
sns.scatterplot(x=df['Age'],y=df['EstimatedSalary'],data=df,hue='Purchased',palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=0.3,random_state=0)
x_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd # Make sure pandas is imported

scaler=MinMaxScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)

# Convert x_train_scaled and x_test_scaled to Pandas DataFrames
x_train_scaled = pd.DataFrame(x_train_scaled, columns=[['Age', 'EstimatedSalary']]) # Use a list of column names
x_test_scaled = pd.DataFrame(x_test_scaled, columns=[['Age', 'EstimatedSalary']]) # Use a list of column names

np.round(x_train.describe(), 1)
```

```python
np.round(x_train_scaled.describe(), 1)
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.scatter(x_train['Age'],x_train['EstimatedSalary'])
ax1.set_title("Before scaling")
ax2.scatter(x_train_scaled['Age'],x_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After scaling")
plt.show  ()
```
